```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2-large')
set_seed(42)
generator("Orangutans are very interesting apes because", max_length=30, num_return_sequences=5)
```

```python
context  = '''Hello how are you
i'm pretty fine
Thats good to know 
'''
```

```python
context
```

```python
while True:
  output = generator(context, max_length=len(context) + 30, num_return_sequences=1)[0]['generated_text'].split(context)[1].split('\n')[0]
  print(output)
  context += output + '\n'
  inputText = input('>>')
  context += inputText + '\n'
```

```python
from transformers import MarianMTModel, MarianTokenizer

tokenizer_en_t_es = MarianTokenizer.from_pretrained(
   "Helsinki-NLP/opus-mt-en-ROMANCE"
)
model_en_t_es = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-en-ROMANCE")


tokenizer_es_t_en = MarianTokenizer.from_pretrained(
    "Helsinki-NLP/opus-mt-ROMANCE-en"
)
model_es_t_en = MarianMTModel.from_pretrained("Helsinki-NLP/opus-mt-ROMANCE-en")
```

```python
def eng_to_spanish(text):
  src_text = [">>es<< {}".format(text)]
  translated = model_en_t_es.generate(
      **tokenizer_en_t_es.prepare_seq2seq_batch(src_text, return_tensors="pt")
  )
  tgt_text = [
      tokenizer_en_t_es.decode(t, skip_special_tokens=True)
      for t in translated
  ]

  return tgt_text[0]
```

```python
def esp_to_english(text):
  src_text = [">>es<< {}".format(text)]
  translated = model_es_t_en.generate(
      **tokenizer_es_t_en.prepare_seq2seq_batch(src_text, return_tensors="pt")
  )
  tgt_text = [
      tokenizer_es_t_en.decode(t, skip_special_tokens=True)
      for t in translated
  ]

  return tgt_text[0]
```

```python
eng_to_spanish('hello world')
```

```python
while True:
  output = '\n'
  while output == '\n':
    output = generator(context, max_length=len(context) + 30, num_return_sequences=1)[0]['generated_text'].split(context)[1].split('\n')[0]
  print(eng_to_spanish(output))
  context += output + '\n'
  inputText = esp_to_english(input('>>'))
  context += inputText + '\n'
```
